```python
import datetime
today = datetime.datetime.now()
today = str(today).split()[0].split('-')
tod = '-'.join(today)
date = '. '.join(today)
today = f'{int(today[0])}. {int(today[1])}. {int(today[2])}'
print(today)
```

```python
import pandas as pd
from utils import check
import gspread


gc = gspread.authorize(creds)
worksheet = gc.open(form_name).sheet1

rows = worksheet.get_all_values()
df_response = pd.DataFrame.from_records(rows[1:], columns = rows[0])

test = gc.open('출석체크')
test_sheet = test.sheet1.get_all_values()
df_test = pd.DataFrame(test_sheet[1:], columns=test_sheet[0])

date = test.sheet1.find(date)

alphabet = 'ABCDEFGHIJKLMNOPQRSTUVXYZ'
col = alphabet[date.col - 1]
rs = 3
re = len(df_test) + 3

final, late = check(df_response, today, deadline, df_test, tod)
test.sheet1.update(f'{col}{rs}:{col}{re}', final)
```

```python
for later in late:
  print(later)
```
